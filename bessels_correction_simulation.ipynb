# **Variance in Normally Distributed Population**

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate population: normal distribution with mean 50, std 10, size 50000
population = np.random.normal(50, 10, 50000)

# True population variance
true_var = np.var(population, ddof=0)

# Sample sizes
sample_sizes = []

for i in range(10, 100, 1):
  sample_sizes.append(i)


# Number of replications per sample size
num_reps = 500

# experimenting with different divisors to calculate variance
divisors = ['n-4', 'n-3', 'n-2', 'n-1', 'n', 'n+1', 'n+2']
results = {div: [] for div in divisors}

for n in sample_sizes:
    for div in divisors:
        sample_vars = []
        for _ in range(num_reps):
            sample = np.random.choice(population, size=n, replace=False)
            sample_mean = np.mean(sample)
            sum_squared_diff = np.sum((sample - sample_mean) ** 2)
            if div == 'n-4':
                k = n - 4
            elif div == 'n-3':
                k = n - 3
            elif div == 'n-2':
                k = n - 2
            elif div == 'n-1':
                k = n - 1
            elif div == 'n':
                k = n
            elif div == 'n+1':
                k = n + 1
            elif div == 'n+2':
                k = n + 2
            sample_var = sum_squared_diff / k
            sample_vars.append(sample_var)
        results[div].append(np.mean(sample_vars))

# Print results
print(f"True Population Variance: {true_var:.4f}")
print("\nSample Size | Avg Var (n-4) | Avg Var (n-3) | Avg Var (n-2) | Avg Var (n-1) | Avg Var (n) | Avg Var (n+1) | Avg Var (n+2)")
print("-" * 80)
for i, n in enumerate(sample_sizes):
    print(f"{n:11d} | {results['n-4'][i]:13.4f} | {results['n-3'][i]:13.4f} | {results['n-2'][i]:13.4f} | {results['n-1'][i]:13.4f} | {results['n'][i]:13.4f} | {results['n+1'][i]:13.4f} | {results['n+2'][i]:13.4f}")

# Plot
plt.figure(figsize=(12, 7))
for div in divisors:
    plt.plot(sample_sizes, results[div], label=f'Avg Est Var (div={div})')
plt.axhline(y=true_var, color='r', linestyle='--', label='True Population Variance')
plt.xlabel('Sample Size')
plt.ylabel('Variance')
plt.title('Average Estimated Variance by Divisor vs. Sample Size')
plt.legend()
plt.grid(True)
plt.show()
```
